```python
img_tri = cv2.imread('./Computer-Vision-with-Python/DATA/dog_backpack.jpg')
img_tri = cv2.cvtColor(img_tri, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_tri)
```

```python
points = np.array([[230, 700], [400, 360], [600, 700]], np.int32)

# Reshape the points to the required shape for polylines
points = points.reshape((-1, 1, 2))

# Define the color (BGR format) and thickness
color = (0, 0, 255)  # Blue color
thickness = 10  # Thickness of the triangle edges

# Draw the triangle using cv2.polylines
cv2.polylines(img_tri, [points], isClosed=True, color=color, thickness=thickness)
plt.imshow(img_tri)
```

```python
cv2.fillPoly(img_tri, [points], color)
plt.imshow(img_tri)
```

```python
# Create a function based on a CV2 Event (Left button click)
def draw_circle(event,x,y,flags,param):
    if event == cv2.EVENT_RBUTTONDOWN:
        cv2.circle(img,(x,y),70,(0,0,255),6)
        

# Create a black image
img = cv2.imread('./Computer-Vision-with-Python/DATA/dog_backpack.jpg')
# This names the window so we can reference it 
cv2.namedWindow(winname='Dog')
# Connects the mouse button to our callback function
cv2.setMouseCallback('Dog',draw_circle)

while True: #Runs forever until we break with Esc key on keyboard
    # Shows the image window
    cv2.imshow('Dog',img)
    if cv2.waitKey(20) & 0xFF == 27:
        break

cv2.destroyAllWindows()
```
